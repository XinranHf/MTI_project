```python
# Import necessary libraries
import numpy as np
import sys
sys.path.append('../utils/')
sys.path.append('../src/')

from initial_param_SPA import initialize_parameters
from SPA import SPA
from plot_RESULT import plot_RESULT
from ISNR import ISNR
from skimage.metrics import structural_similarity as ssim

# Load all parameters
params = initialize_parameters()

# Unpack parameters
X = params['X']
Y = params['Y']
N = params['N']
sigma = params['sigma']
rho = params['rho']
beta = params['beta']
alpha = params['alpha']
N_MC = params['N_MC']
N_bi = params['N_bi']
y_signal = params['y_signal']
Hmat = params['Hmat']
invQ = params['invQ']
M = params['M']
```

## Load workspace variables and launch SPA algorithm
Load workspace variables (defined in `../utils/initial_param_SPA.py`) and launch SPA algorithm.

```python
# Run SPA algorithm
X_MC, Z_MC, U_MC = SPA(y_signal, Hmat, sigma, rho, beta, alpha, N, M, invQ, N_MC)
```

## Display SNR, MSE and SSIM
Display SNR, MSE and SSIM associated to the MMSE estimator of x.

```python
# Calculate ISNR and MSE
isnr, mse = ISNR(X, Y, X_MC, N_bi)

# Calculate SSIM
X_MMSE = np.mean(X_MC[:, :, N_bi:], axis=2)
SSIM = ssim(X_MMSE, X, data_range=X.max() - X.min())

print(f'ISNR: {isnr:.4f} dB')
print(f'MSE: {mse:.4f}')
print(f'SSIM: {SSIM:.4f}')
```

## Plot the results
Display the original image, degraded image, and reconstructed results.

```python
# Plot results
plot_RESULT(Y, X, X_MC, Z_MC, U_MC, N_bi, N_MC, N)
```
